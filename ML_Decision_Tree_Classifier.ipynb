# ML_Decision_Tree_Classifier

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\yarra\OneDrive\Desktop\Data sets\Salaries_Decision Tree.csv")
df
```

```python
x=df.iloc[:,0:3]
x
```

```python
y=df.iloc[:,3:]
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_com=LabelEncoder()
l_job=LabelEncoder()
l_deg=LabelEncoder()
```

```python
x["l_company"]=l_com.fit_transform(x["company"])
x["l_job"]=l_job.fit_transform(x["job"])
x["l_degree"]=l_deg.fit_transform(x["degree"])
```

```python
x
```

```python
x=x.iloc[:,3:6]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.70)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

```python
dtc.predict(xtest)
```

```python
dtc.score(xtest,ytest)
```
