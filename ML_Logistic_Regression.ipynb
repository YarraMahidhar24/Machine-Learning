# ML_Logistic_Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv(r"C:\Users\yarra\OneDrive\Desktop\Data sets\Iris.csv")
df
```

```python
x=df.iloc[:,1:5]
x
```

```python
y=df.iloc[:,5:]
y
```

# SPLITING dATA

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8) # or test_size=0.2
```

```python
xtrain  # training data inputs
```

```python
ytrain # training data output
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
xtest   # testing data inputs
```

```python
ytest    # testing data output
```

```python
ans=model.predict(xtest)
ans
```

```python
model.score(xtest,ytest)
```

```python
model.predict([[2,3.5,2,4]])
```

# Confussion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(ans,ytest)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
```
