# ML_Polynomial_Regression

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\yarra\OneDrive\Desktop\Data sets\Salaries.csv")
df
```

```python
x=df.iloc[:,1:2]
x   # single input linear
```

```python
y=df.iloc[:,2:]
y
```

```python
from sklearn.preprocessing import PolynomialFeatures 
```

```python
poly=PolynomialFeatures(degree=2) # ASSIGING DEGREE
```

```python
x_poly=poly.fit_transform(x)
x_poly   # converted into polynomial ax**2+bx+c
```

```python
from sklearn.linear_model import LinearRegression
poly_model=LinearRegression()
```

```python
poly_model.fit(x_poly,y)
```

```python
pred=int(input("ENTER THE LEVEL TO PREDICT THE SALERY = "))
```

```python
pred_model=poly.fit_transform([[pred]])
```

```python
poly_model.predict(pred_model)
```

```python
plt.scatter(x,y,color='blue')
plt.plot(x,poly_model.predict(poly.fit_transform(x)),color='r')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polynomial Regression")
```
