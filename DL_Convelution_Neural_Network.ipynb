```python
import cv2
import matplotlib.pyplot as plt
```

## CV2 file data importing to use the cnn

```python
face=cv2.CascadeClassifier(r"C:\Users\yarra\anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")
```

```python
img=cv2.imread(r"C:\Users\yarra\OneDrive\Desktop\Data sets\images.jpeg.jpg")
im=cv2.imread(r"C:\Users\yarra\OneDrive\Desktop\Data sets\Familyimg.png")
plt.imshow(img)
```

```python
plt.imshow(im)
```

## image convertion

```python
gray1=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray1)
```

```python
gray2=cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
plt.imshow(gray2)
```

## face detection

```python
co=face.detectMultiScale(gray1,scaleFactor=1.1,minNeighbors=1,minSize=(100,100))
co
```

```python
co1=face.detectMultiScale(gray2,scaleFactor=1.1,minNeighbors=4,minSize=(100,100))
co1
```

## drowing a box to the identified face

```python
img1=cv2.imread(r"C:\Users\yarra\OneDrive\Desktop\Data sets\images.jpeg.jpg")
for (x,y,w,h) in co:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,0,255),3)
```

```python
im=cv2.imread(r"C:\Users\yarra\OneDrive\Desktop\Data sets\Familyimg.png")
for (x,y,w,h) in co1:
    cv2.rectangle(im,(x,y),(x+w,y+h),(255,0,0),5)
```

## displaying the image after box drawing

```python
output=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(output)
```

```python
output1=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(output1)
```
