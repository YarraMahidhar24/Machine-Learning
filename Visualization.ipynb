## Data visualization

1.Data visualization is the graphical representation of information and data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
roll=[1,2,3,4,5]
marks=[9,27,36,89,69]
```

```python
plt.plot(roll,marks)
```

```python
df100=pd.DataFrame({"Roll":roll,"Marks":marks})
```

```python
plt.scatter(df100["Roll"],df100["Marks"])
```

```python
plt.scatter(df100["Roll"],df100["Marks"])
plt.xlabel("Roll")
plt.ylabel("Marks")
plt.title("Roll Vs Marks")
```

```python
plt.scatter(df100["Roll"],df100["Marks"],marker='x')
plt.xlabel("Roll")
plt.ylabel("Marks")
plt.title("Roll Vs Marks")
```

```python
clrs=['r','y','g','b','c']
plt.scatter(df100["Roll"],df100["Marks"],marker='*',c=clrs)
plt.xlabel("Roll")
plt.ylabel("Marks")
plt.title("Roll Vs Marks")
```

```python
size=[100,200,300,400,500]
clrs=['r','y','g','b','c']
plt.scatter(df100["Roll"],df100["Marks"],marker='*',c=clrs,s=size[::-1])
plt.xlabel("Roll")
plt.ylabel("Marks")
plt.title("Roll Vs Marks")
```

```python
sns.lineplot(x="Roll",y="Marks",data=df100)
```

```python
clrs=['r','y','g','b','c']
plt.bar(df100["Roll"],df100["Marks"],color=clrs)
```

```python
plt.bar(df100["Roll"],df100["Marks"],color=clrs,width=0.5)
```

```python
plt.bar(df100["Roll"],df100["Marks"],color=clrs,width=0.5,alpha=0.5)
```
